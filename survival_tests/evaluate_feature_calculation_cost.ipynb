```python
from aslib_scenario import ASlibScenario
import os
```

```python
feature_cost_data = {}
for scenario_name in ['CSP-2010','QBF-2016', 'ASP-POTASSCO','SAT12-HAND', 'MAXSAT15-PMS-INDU','SAT12-INDU','CPMP-2015',]:
    scenario = ASlibScenario()
    scenario.read_scenario(os.path.join('data', 'aslib_data-master', scenario_name))
    feature_cost_data[scenario_name] = scenario.feature_cost_data

    
```

```python
for x,y in feature_cost_data.items():
    if y is not None:
    	print(x, 'has cost data')
```
