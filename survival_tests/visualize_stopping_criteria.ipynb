```python
from fetch_dataframe_from_server import fetch_dataframe_from_server
from matplotlib import pyplot as plt
import numpy as np
```

```python
sql_statement = """SELECT scenario_name, approach,metric,impact_factor,ranking_error,borda_score,stopping_criteria,AVG(result) AS result
FROM `final_stopping_fixed`
GROUP BY scenario_name, approach,metric,impact_factor,ranking_error,borda_score,stopping_criteria"""
sql_baselines = """SELECT scenario_name, approach, metric, AVG(result) as result FROM `baselines` WHERE approach = 'sbs' GROUP BY scenario_name, approach, metric"""

df = fetch_dataframe_from_server(sql_statement)
df = df[df["approach"] == "BinaryDecisionTree"]

df_baseliens = fetch_dataframe_from_server(sql_baselines)
df
```

```python
for metric in df["metric"].unique():
    fig, axs = plt.subplots(4, 2,)
    axs = axs.flatten()
    fig.delaxes(axs[-1])
    fig.tight_layout()
    for scenario_number, scenario_name in enumerate(df["scenario_name"].unique()):
        fig.set_size_inches(30, 20, forward=True)
        rows = list()
        fig.suptitle("Stopping Criteria - Ablation Study - " + metric , fontsize=40)
        axs[scenario_number].figsize = (10, 5)
        axs[scenario_number].patch.set_edgecolor("black")
        axs[scenario_number].patch.set_linewidth("2")
        axs[scenario_number].patch.set_edgecolor("black")
        axs[scenario_number].patch.set_linewidth("2")
        for stopping_criterion in sorted(df["stopping_criteria"].unique()):
            rows.append(list(df[df["scenario_name"] == scenario_name][df["metric"] == metric][df["stopping_criteria"] == stopping_criterion]['result'])[0])
        rows.append(list(df_baseliens[df_baseliens['scenario_name'] == scenario_name][ df_baseliens['metric'] == metric]['result'])[0])
        rows = rows[0:11] + rows[-17:]
        subjects = []
        for element in list(sorted(df["stopping_criteria"].unique())[0:11]) + list(sorted(df["stopping_criteria"].unique())[-16:]):
            if not 'None' in element:
                val:float = round(float(element.split('_')[-1]), 3)
                if val.is_integer():
                    val = subjects.append(element[0] + ' ' + str(int(val)))
                else:
                    subjects.append(element[0] + ' '+ str(val))
            else:
                subjects.append('all')
        subjects.append('sbs')
        rows = np.array(rows)
        bottom = np.min(rows) - 1/50* np.max(rows)
        rows = rows - bottom
        rects = axs[scenario_number].bar(subjects, rows, bottom=bottom, color = ['darkslateblue'] * 11 +  ['purple']*6+ ['darkred'] * 10 + ['darkorange'], alpha = 0.8,)
        
        axs[scenario_number].tick_params(
            axis='x',          # changes apply to the x-axis
            which='both',      # both major and minor ticks are affected
            bottom=True,      # ticks along the bottom edge are off
            top=False,         # ticks along the top edge are off
            labelbottom=True,
            labelrotation=90,
            labelsize=25,)
        if metric == 'par10':
            metric_name_graph = 'Par10'
        if metric == 'performanceRegret':
            metric_name_graph = 'Performance Regret'
        if metric == 'number_unsolved_instances_False':
            metric_name_graph = 'Percentage of Unsolved Instances'
        if metric == 'NDCG':
            metric_name_graph = 'NDCG'
        axs[scenario_number].set_title(scenario_name, fontsize=20)
        axs[scenario_number].tick_params(axis="both", which="major", labelsize=15)
        axs[scenario_number].tick_params(axis='y', which='major', labelsize=25)
        patterns = ['//'] * 11 + ['\\\\']*6 + [ '||'] * 10 + [ '-']
        for pattern, bar in zip(patterns, axs[scenario_number].patches):
            bar.set_hatch(pattern)
        fig.legend([rects[0], rects[-15], rects[-2], rects[-1]],['loss under threshold', 'max depth', 'same ranking', 'sbs'], bbox_to_anchor=(0.21, -0.252, 0.5, 0.5), prop = {'size': 30})
        #plt.show()
    plt.savefig(f"stopping_criteria_{metric}.png")
    plt.show()
    plt.clf()
```
