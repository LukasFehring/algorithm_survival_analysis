```python
from fetch_dataframe_from_server import compare_losses_and_lambda
from matplotlib import pyplot as plt
```

```python
starting_SQL_trees = """SELECT scenario_name, approach,metric, impact_factor, ranking_error, borda_score, stopping_criteria, AVG(result) AS result FROM `do_preprocessing`
WHERE approach LIKE "Bin%"
AND ranking_error = '{}'
AND metric = '{}'
AND scenario_name LIKE "{}"
GROUP BY scenario_name, approach, metric, impact_factor, ranking_error, borda_score, stopping_criteria"""

sql = """SELECT scenario_name, approach, metric, impact_factor, ranking_error, borda_score, stopping_criteria, AVG(result) as result
FROM `comparison_to_others`
WHERE approach LIKE 'Binary%'
AND ranking_error = '{}'
AND metric = '{}'
AND scenario_name LIKE '{}'
AND stopping_criteria = '{}'
GROUP BY scenario_name, approach, metric, impact_factor, ranking_error, borda_score, stopping_criteria"""
```

```python
scenario_names = ['CSP-2010','QBF-2016', 'ASP-POTASSCO','SAT12-HAND', 'MAXSAT15-PMS-INDU','SAT12-INDU','CPMP-2015',]
ranking_losses = ['spearman_rank_correlation', 'spearman_rankk_correlation_no_normalisation']
ranking_loss_names = ['Spearman Rank Correlation - [0,1] scaled', 'Spearman Rank Correlation - not [0,1] scaled']
baseline_approaches = ['sbs']
for scenario_name in scenario_names:
    for metric_name in ['par10',  'performanceRegret','number_unsolved_instances_False', 'NDCG']:
        metric_name_graph = None
        sql_trees = [starting_SQL_trees.format(ranking_loss, metric_name, scenario_name) for ranking_loss in ranking_losses]
        if metric_name == 'par10':
            metric_name_graph = 'Par10'
        if metric_name == 'performanceRegret':
            metric_name_graph = 'Performance Regret'
        if metric_name == 'number_unsolved_instances_False':
            metric_name_graph = 'Percentage of Unsolved Instances'
        if metric_name == 'NDCG':
            metric_name_graph = 'NDCG'
        title = f"Comparison of Hybrid Binary Decision Trees With, or Without Modified Specification on {scenario_name} with {metric_name_graph}"
        compare_losses_and_lambda(sql_trees, ranking_loss_names,r"$\lambda$", metric_name_graph,title, [] )
        plt.show()
```
